```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_agendamentos = {
    'Servio': ['Consulta Veterinria', 'Banho', 'Tosa'],
    'Agendamentos': [75, 150, 130]
}

df_agendamentos = pd.DataFrame(data_agendamentos)

data_cancelamentos = {
    'Motivo': ['Cliente Desistiu', 'Problema de Sade', 'Outro'],
    'Cancelamentos': [10, 5, 2]
}

df_cancelamentos = pd.DataFrame(data_cancelamentos)

# Configuraes gerais do grfico
sns.set(style="whitegrid")

fig, axs = plt.subplots(2, 1, figsize=(14, 10))

# Grfico de agendamentos por servio
sns.barplot(x='Servio', y='Agendamentos', data=df_agendamentos, ax=axs[0], palette="Blues_d")
axs[0].set_title('Agendamentos por Servio', fontsize=16)
axs[0].set_xlabel('Servio', fontsize=14)
axs[0].set_ylabel('Nmero de Agendamentos', fontsize=14)

# Grfico de cancelamentos por motivo
sns.barplot(x='Motivo', y='Cancelamentos', data=df_cancelamentos, ax=axs[1], palette="Reds_d")
axs[1].set_title('Cancelamentos por Motivo', fontsize=16)
axs[1].set_xlabel('Motivo', fontsize=14)
axs[1].set_ylabel('Nmero de Cancelamentos', fontsize=14)

# Adicionando informaes no dashboard
fig.suptitle('Dashboard de Agendamentos e Cancelamentos', fontsize=20)

# Anotaes e insights
agendamentos_total = df_agendamentos['Agendamentos'].sum()
cancelamentos_total = df_cancelamentos['Cancelamentos'].sum()
agendamentos_mensal = agendamentos_total / 3  # exemplo simples
cancelamentos_mensal = cancelamentos_total / 3  # exemplo simples

insights_text = (f'Total de Agendamentos: {agendamentos_total}\n'
                 f'Total de Cancelamentos: {cancelamentos_total}\n'
                 f'Mdia Mensal de Agendamentos: {agendamentos_mensal:.2f}\n'
                 f'Mdia Mensal de Cancelamentos: {cancelamentos_mensal:.2f}\n')

# Adicionando o texto de insights ao grfico
plt.figtext(0.5, -0.05, insights_text, ha="center", fontsize=12, bbox={"facecolor":"lightgrey", "alpha":0.5, "pad":5})

plt.tight_layout(rect=[0, 0.05, 1, 0.95])
plt.show()
```
